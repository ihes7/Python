```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# البيانات
data = pd.DataFrame({
  "Department": ["Sales", "Marketing", "IT", "HR", "Finance", "Operations", "Logistics", "R&D", "Legal", "Admin"],
  "Revenue": [500000, 350000, np.nan, 150000, 400000, 450000, 300000, -10000, "", 250000], 
  "Expenses": [200000, np.nan, 100000, 80000, 150000, "", 180000, 90000, 120000, 220000], 
  "Profit": [300000, 170000, 100000, 70000, 250000, 200000, np.nan, -20000, 130000, 30000], 
  "Budget Allocation": [550000, 370000, 220000, "", 420000, 460000, 320000, 0, 210000, np.nan], 
  "Variance": [50000, 20000, np.nan, 30000, 20000, 10000, -5000, 15000, 0, 8000], 
  "Financial Risk": ["Low", "Medium", "Low", "High", "", "Medium", "High", "Low", "Medium", "High"] 
})
data
```

```python
#Data Cleaning
data.replace("", np.nan, inplace=True)
data
```

- Data Visualization

```python
plt.figure(figsize=(8, 6), facecolor='#F8F8F8')
x = sns.color_palette(["#F7879A", "#FFC0CB"])  
sns.heatmap(data.isnull(), cmap=sns.color_palette(x, as_cmap=True), cbar=False, linewidths=0.5)
plt.title("Missing Values")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 6),facecolor='#F8F8F8')
sns.violinplot(x=data['Profit'], color='#F7879A')
plt.title("Revenue Distribution")
plt.xlabel("Revenue")
plt.grid(axis='x', linestyle='--') 
plt.show()
```

```python
plt.figure(figsize=(8, 6),facecolor='#F8F8F8')
sns.scatterplot(x=range(len(data)), y=data['Profit'], color="#F7879A")
plt.title("Profit Values")
plt.xlabel("Index")
plt.ylabel("Profit")
plt.grid(True,linestyle='--')
plt.show()
```

```python
plt.figure(figsize=(8, 6), facecolor='#F8F8F8')
colors = ['#F7879A', 'pink', '#FF85B2', '#FFA6C9', '#E75480', '#FFD1DC']
c = sns.color_palette(colors)
sns.boxplot(data=data.select_dtypes(include=[np.number]), palette=c)
plt.title("Boxplot for Outlier Detection")
plt.xticks(rotation=45)
plt.show()
```

- Data Cleaning

```python
data = data.dropna(subset=['Department', 'Revenue', 'Expenses', 'Profit'])
data
```

```python
data['Profit'] = data['Profit'].apply(lambda x: x if x >= 0 else np.nan)
data
```

```python
data['Profit'].fillna('0', inplace=True)
data
```

```python
MV = data["Budget Allocation"].mean(skipna=True)
data["Budget Allocation"] = data["Budget Allocation"].apply(lambda x: x if pd.notna(x) and x > 0 else MV)
# تقليل الأرقام العشرية إلى خانة واحدة فقط
data["Budget Allocation"] = data["Budget Allocation"].round(1)
data
```

```python
data['Variance'] = data['Variance'].apply(lambda x: x if x >= 0 else np.nan)
data
```

```python
FR = {'Low': 0, 'Medium': 1, 'High': 2}
data['Financial Risk'] = data['Financial Risk'].map(FR) 
data
```

```python
data['Financial Risk'].fillna('-1', inplace=True)
data
```

```python
data = data.drop_duplicates()
data
```

```python
data=data.drop(data[data['Revenue']<0].index)
data
```

```python
data['Financial Risk'] = pd.to_numeric(data['Financial Risk'], errors='coerce')
data['Profit'] = pd.to_numeric(data['Profit'], errors='coerce')
data
```

```python
data['Financial Risk'].fillna(0, inplace=True)
data['Profit'].fillna(0, inplace=True)
data
```

```python
data['Financial Risk'] = data.apply(
    lambda row: 0 if row['Financial Risk'] < 0 and row['Profit'] >= 100000 else
                2 if row['Financial Risk'] < 0 and row['Profit'] < 100000 else
                row['Financial Risk'],axis=1 )
data
```
