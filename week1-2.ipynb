#### Company's Monthly Financial Performance

```python
import pandas as pd
import matplotlib.pyplot as plt

#  DataFrame        
data = pd.DataFrame({
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Revenue': [50000, 52000, 48000, 51000, 53000, 55000, 60000, 62000, 58000, 57000, 59000, 61000],  # 
    'Expenses': [30000, 32000, 31000, 30000, 34000, 33000, 35000, 36000, 34000, 32000, 31000, 30000],  # 
    'Profit': [20000, 20000, 17000, 21000, 19000, 22000, 25000, 26000, 24000, 25000, 28000, 31000],  # 
    'Operations': [15000, 16000, 15500, 15200, 16500, 15800, 17000, 17500, 16900, 16000, 15800, 15700],  #  
    'Marketing': [5000, 5200, 5100, 4800, 5300, 5500, 6000, 6200, 5800, 5700, 5900, 6100],  #  
    'R&D': [8000, 8200, 7900, 8500, 8700, 8600, 9000, 9500, 9100, 9200, 8900, 8800]  #  
})
fig,x = plt.subplots(figsize=(8,6),facecolor='#F8F8F8')
x.set_facecolor('lightgray')
plt.plot( data['Month'],data['Revenue'], marker='o', linestyle='--', label='Revenue', color='#F8F8F8')
plt.plot( data['Month'],data['Expenses'], marker='o', linestyle='--', label='Expenses', color='#F7879A')
plt.plot( data['Month'],data['Profit'], marker='o', linestyle='--',label='Profit',  color='darkred')



plt.xlabel('Month')
plt.ylabel('Total Sales ($)')
plt.title('Sales Over Years')
plt.grid(True)
plt.legend()
plt.show()

#   5   
print(data.head())
```

#### Financial Distribution Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
#    
month_data = data[data['Month'] == 'Aug'].iloc[0]
values = [month_data['Revenue'], month_data['Expenses'], month_data['Profit'], month_data['Operations'], month_data['Marketing'], month_data['R&D']]
labels = ['Revenue', 'Expenses', 'Profit', 'Operations', 'Marketing', 'R&D']
colors = ['#F7879A', 'pink', '#FF85B2', '#FFA6C9', '#E75480', '#FFD1DC']

fig= plt.subplots(figsize=(8,6),facecolor='#F8F8F8')
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title(f'Financial Distribution for {month_data["Month"]}')
plt.legend()
plt.show()

```

#### Monthly Expense

```python
fig,x= plt.subplots(figsize=(8,6),facecolor='#F8F8F8')
x.set_facecolor('lightgray')
plt.barh(data['Month'], data['Expenses'], color='#E75480', alpha=0.8)
plt.title('Monthly Expense')
plt.xlabel('Expenses ($)')
plt.ylabel('Month')
plt.grid(axis='x', linestyle='--')
plt.legend('E')
plt.show()
```

#### Monthly Revenue

```python
fig,x= plt.subplots(figsize=(8,6),facecolor='#F8F8F8')
x.set_facecolor('lightgray')
plt.plot(data['Month'], data['Revenue'], marker='o', linestyle='--', color='darkred',alpha=0.8, label='Revenue')  
plt.title('Monthly Revenue')
plt.xlabel('Month')
plt.ylabel('Revenue ($)')
plt.legend()
plt.grid(True,linestyle='--')
plt.show()
```

#### Monthly Profit

```python
fig,x= plt.subplots(figsize=(8,6),facecolor='#F8F8F8')
x.set_facecolor('lightgray')
plt.bar(data['Month'], data['Profit'], color='#E75480', alpha=0.8,label='Profit') 
plt.title('Monthly Profit')
plt.xlabel('Month')
plt.ylabel('Profit ($)')
plt.grid(axis='y', linestyle='--') 
plt.legend()
plt.show()
```
